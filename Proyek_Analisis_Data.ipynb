### Cleaning Data

Mengubah tipe data kolom dteday di ```#hour_df``` menjadi datetime

````python
datetime_columns = ['dteday']

for columns in datetime_columns:
  hour_df[columns] = pd.to_datetime(hour_df[columns])
````

````python
hour_df.info()
````

Mengubah tipe data kolom dteday di ```#day_df``` menjadi datetime

````python
datetime_columns = ['dteday']

for columns in datetime_columns:
  day_df[columns] = pd.to_datetime(day_df[columns])
````

Mengubah value season dari 1-4 menjadi Winter, Spring, Summer, Fall

````python
day_df["season"] = day_df["season"].astype(str)
````

````python
season_mapping = {'1': 'Winter', '2': 'Spring', '3': 'Summer', '4': 'Fall'}
day_df['season'] = day_df['season'].map(season_mapping)
````

## Visualization & Explanatory Analysis

### Pertanyaan 1: Jam berapa peminjam paling banyak?

````python
result = hour_df.groupby(by="hr").instant.nunique().sort_index()

plt.figure(figsize=(10, 6))
plt.plot(result.index, result.values, marker='o', linewidth=2, color="#82CAFF")
plt.title("Jumlah Peminjam pada Setiap Jam")
plt.xlabel("Jam")
plt.ylabel("Jumlah Peminjam")
plt.xticks(result.index, fontsize=10)
plt.yticks(fontsize=10)
plt.show()
````

### Pertanyaan 2: Musim apa peminjam sepeda terbanyak?

````python
result = day_df.groupby(by="season").instant.nunique().sort_index().reset_index()

sns.barplot(
    y="instant",
    x="season",
    data=result,
    palette=["#D3D3D3", "#D3D3D3", "#72BCD4", "#D3D3D3"],
)

plt.title("Jumlah Peminjam pada Setiap Musim (per hari)")
plt.xlabel("Musim")
plt.ylabel("Jumlah Peminjam")
plt.xticks(result.index, fontsize=10)
plt.yticks(fontsize=10)
plt.show()
````
